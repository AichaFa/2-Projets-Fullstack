# Build a quiz - Part 2

Below is the code we wrote yesterday, that asks three questions to the user and allows up to 3 wrong answers :

```python
# --- 1. Structure de Données (DRY) ---
# Toutes les questions sont stockées dans une liste, facile à étendre.
QUESTIONS_DU_QUIZ = [
    {
        "question": "Combien de fois la France a gagné la coupe du monde ?",
        "reponse_attendue": "2"
    },
    {
        "question": "Quand a été fondé Apple ? (AAAA)",
        "reponse_attendue": "1976"
    },
    {
        "question": "Qui a fondé SpaceX ? (Nom Prénom)",
        "reponse_attendue": "elon musk"
    }
]

# --- 2. Fonction pour Gérer la Logique Répétitive ---
def poser_question(question, reponse_attendue, nb_de_chances_restantes): # Fonction qui pose une question et gère les tentatives
  
    print() # Ajoute un saut de ligne pour séparer la question
    reponse_utilisateur = input(f"{question} ") # Première demande de réponse
    
    # Boucle tant que la réponse est FAUSSE (gestion des tentatives grâce au 'while')
    while reponse_utilisateur.strip().lower() != reponse_attendue.lower():
        nb_de_chances_restantes -= 1 # Décompte une chance

        if nb_de_chances_restantes <= 0: # Si plus de chances restantes
            print("Dommage ! Il te reste 0 chances. Fin de cette question.")
            return 0  # Échec définitif
        
        # S'il reste des chances, on informe l'utilisateur
        print(f"Dommage ! Il te reste {nb_de_chances_restantes} chances.")

        # On repose la question
        reponse_utilisateur = input(f"{question} ")

    # Si on sort de la boucle, la réponse était bonne
    print(f"{question} {reponse_attendue} - Bonne réponse !")
    return nb_de_chances_restantes # Retourne les chances restantes

# --- 3. Programme Principal (Boucle DRY) ---

NB_CHANCES_INITIALES = 3 # Nombre initial de chances
nb_de_chances = NB_CHANCES_INITIALES # Initialisation des chances
nb_reponses_correctes = 0 # Compteur de bonnes réponses

print(f"Voici notre quiz, tu as {NB_CHANCES_INITIALES} chances !")

# La boucle 'for' parcourt toutes les questions sans répéter de code
for item in QUESTIONS_DU_QUIZ:
    if nb_de_chances > 0:
        # L'appel de fonction gère toute la logique de la question
        nouvelles_chances = poser_question(
            question=item["question"],
            reponse_attendue=item["reponse_attendue"],
            nb_de_chances_restantes=nb_de_chances
        )

        if nouvelles_chances > 0: # Si la réponse était correcte
            nb_reponses_correctes += 1 # Incrémente le compteur de bonnes réponses
            nb_de_chances = nouvelles_chances # Mise à jour des chances
        else:
            nb_de_chances = 0 # Met à jour les chances à 0 en cas d'échec
            break # Arrête le quiz si l'échec est atteint
            
# --- 4. Affichage du Résultat Final ---
print("\n--- Resultat du Quiz ---")

if nb_reponses_correctes == len(QUESTIONS_DU_QUIZ): 
    # Si 3 bonnes réponses sans perte de chance)
    print(f"Bravo ! Tu as gagné le quiz avec {nb_de_chances} chances restantes !")
else: 
    print(f"Fin du jeu. Tu as obtenu {nb_reponses_correctes} bonnes réponses.")
    if nb_de_chances == 0: # Si l'utilisateur a perdu toutes ses chances
        print("Tu as perdu car tu n'avais plus de chances pour continuer.")
```
