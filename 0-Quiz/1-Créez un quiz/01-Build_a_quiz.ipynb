Let's tackle our first project together. Find a partner and start coding a quiz together! Your program should: 

1. Ask a question to a user via the console 
2. Verify whether the answer is correct or not 
3. If the answer is correct, you should print out: `Good job! This is the right answer`

* Else, it should print out : `Too bad! that is not the correct answer`

4. Ask three questions following this model.



```python
def run_quiz():
    """
    Exécute un petit programme de quiz.
    Contient trois questions et vérifie les réponses de l'utilisateur.
    """
    
    # Définir les questions et les réponses
    questions = [
        {
            "prompt": "Quel est mon prénom ?\n", 
            "answer": "aicha"
        },
        {
            "prompt": "De combien d'heure est une journée ?\n",
            "answer": "24"
        },
        {
            "prompt": "Qui est le PDG actuel de Tesla et SpaceX ?\n",
            "answer": "elon musk"
        }
    ]

    print("Here you have our quiz !\n-------")

    # Parcourir toutes les questions et obtenir les réponses
    for q in questions:
        # Afficher la question
        user_answer = input(q["prompt"])
        
        # *** AJOUT DE LA LIGNE MANQUANTE ***
        # Afficher l'entrée brute de l'utilisateur comme demandé.
        print(f"Your answer : {user_answer}") 
        
        # Préparer les réponses pour une comparaison insensible à la casse
        user_answer_processed = user_answer.lower().strip()
        correct_answer_processed = q["answer"].lower().strip() 
        
        # Afficher la réponse traitée pour la question textuelle
        if "elon musk" in correct_answer_processed: 
            print(f"Your answer after .lower() function : {user_answer_processed}")

        # Vérifier la réponse
        if user_answer_processed == correct_answer_processed:
            print("Good job! This is the right answer\n")
        else:
            print("Too bad! that is not the correct answer\n")

# Démarrer le quiz
if __name__ == "__main__":
    run_quiz()
```

There's a problem with our quizz : if the user has not given the right answer, the program goes to next question. We would rather like to ask again the same question if the answer was wrong.

5. Using a `while` loop, make sure that the same question will be asked again when the answer was wrong

6. Rewrite your code such that this behaviour is applied to all questions.

```python
def run_quiz():
    
    # 1. Définir les questions et les réponses
    # Les réponses sont stockées en minuscules pour faciliter la vérification.
    questions = [
        {"prompt": "Combien a-t-on de mains ? ","answer": "2"},
        {"prompt": "Date de la création d'Apple ? ","answer": "1976"},
        {"prompt": "Qui est le PDG actuel de Tesla et SpaceX ?\nVotre réponse : ","answer": "elon musk"}
    ]
    print("Here you have our quiz !\n-------")

    # 2. Parcourir toutes les questions
    for i, q in enumerate(questions):
        # Afficher le numéro de la question
        print(f"\nQuestion {i + 1}")
        
        correctly_answered = False  # Drapeau pour vérifier si la question a été correctement répondue
        
        # 3. Boucle while : Répéter la question tant que la réponse est incorrecte
        while not correctly_answered:
            
            # Afficher la question et obtenir la réponse de l'utilisateur
            user_answer = input(q["prompt"])
            
            # Afficher l'entrée brute de l'utilisateur
            print(f"Your answer : {user_answer}")
            
            # Préparer les réponses pour une comparaison insensible à la casse et aux espaces
            user_answer_processed = user_answer.lower().strip()
            correct_answer_processed = q["answer"].lower().strip() 
            
            # 4. Vérifier la réponse
            if user_answer_processed == correct_answer_processed:
                print("Good job! This is the right answer\n")
                # Définir le drapeau sur True pour sortir de la boucle while
                correctly_answered = True
            else:
                print("Too bad! That is not the correct answer")

    # 5. Message de fin
    print("\nWell done! You are the quiz winner!")

# Démarrer le quiz
if __name__ == "__main__":
    run_quiz()
```

7. Improve your code : if the user makes 3 wrong answers, the program should stop and print ("Too bad, you lost the game !")


```python
def run_quiz():
    # 1. Définir les questions et les réponses
    # Les réponses sont stockées en minuscules pour faciliter la vérification.
    questions = [
        {"prompt": "Combien a-t-on de mains ? ", "answer": "2"},
        {"prompt": "Date de la création d'Apple ? ", "answer": "1976"},
        {"prompt": "Qui est le PDG actuel de Tesla et SpaceX ?\nVotre réponse : ", "answer": "elon musk"}
    ]
    
    # Initialiser le compteur de vies (mauvaises réponses autorisées)
    lives = 3
    
    print("Welcome to our quizz!")
    print(f"You have {lives} lives.")
    print("-------")

    # 2. Parcourir toutes les questions
    for i, q in enumerate(questions):
        # Vérifier si l'utilisateur a encore des vies avant de poser la question
        if lives <= 0:
            print("\nOh no, you lost the quiz...")
            return  # Arrêter la fonction run_quiz
            
        # Afficher le numéro de la question
        print(f"\nQuestion {i + 1}")
        
        correctly_answered = False  # Drapeau pour vérifier si la question a été correctement répondue
        
        # 3. Boucle while : Répéter la question tant que la réponse est incorrecte et qu'il reste des vies
        while not correctly_answered and lives > 0:
            
            # Afficher la question et obtenir la réponse de l'utilisateur
            # Note: Pour que la sortie corresponde exactement à l'exemple, nous devons 
            # simuler l'affichage de "Votre réponse : " pour la première question 
            # et "Your answer : " avant l'entrée.
            
            # Pour la cohérence avec l'exemple de sortie, l'invite est modifiée ici, 
            # mais elle est conservée dans l'objet question pour l'appel `input()`.
            
            # Afficher le prompt de la question (sans "Votre réponse : " si ce n'est pas la dernière)
            if "Votre réponse : " in q["prompt"]:
                # Si c'est la dernière question avec l'invite incluse
                user_answer = input(q["prompt"])
            else:
                # Autres questions
                user_answer = input(q["prompt"])

            # Afficher l'entrée brute de l'utilisateur pour simuler "Your answer : {user_answer}"
            # L'exemple de sortie montre "Your answer : " suivi de la réponse de l'utilisateur.
            # Cependant, dans une exécution réelle avec `input()`, l'utilisateur tape 
            # APRÈS l'invite. Pour correspondre exactement à l'exemple, nous 
            # devons imprimer la réponse de l'utilisateur *après* l'appel à `input()`.
            # Nous utilisons le même format de ligne que l'exemple.
            print(f"Your answer : {user_answer}")
            
            # Préparer les réponses pour une comparaison insensible à la casse et aux espaces
            user_answer_processed = user_answer.lower().strip()
            correct_answer_processed = q["answer"].lower().strip() 
            
            # 4. Vérifier la réponse
            if user_answer_processed == correct_answer_processed:
                # Affichage adapté pour correspondre au format de l'exemple, qui ne 
                # montre pas de message de succès "Good job!"
                # print("Good job! This is the right answer\n") 
                correctly_answered = True
            else:
                lives -= 1  # Décrémenter les vies
                
                if lives > 0:
                    print(f"Sorry, you have {lives} chances left")
                else:
                    # Le message de défaite sera imprimé après la boucle while
                    pass

        # Si la question a été répondue correctement, passer à la suivante
        if correctly_answered:
            # L'exemple de sortie ne montre pas de saut de ligne supplémentaire 
            # après une bonne réponse, on ne fait donc rien ici.
            pass
            
    # 5. Message de fin
    if lives > 0:
        print("\nWell done! You are the quiz winner!")
    else:
        # Ce cas est géré par le 'return' dans la boucle for, 
        # mais on l'ajoute ici pour le cas où l'utilisateur perd 
        # exactement à la fin de la dernière question.
        print("Sorry, you have 0 chances left")
        print("\nOh no, you lost the quiz...")

# Démarrer le quiz
if __name__ == "__main__":
    run_quiz()
```
