### 1.1 Importation des bibliothèques / librairies et geocodage
Récupération des coordonnées géographiques :
<br> La première étape consiste à obtenir les coordonnées géographiques des 35 villes cibles. 
<br> Sans ces coordonnées, nous ne pourrons pas interroger l'API météo ni localiser les hôtels lors de l'étape de scraping.

Méthodologie :
<br>Nous utilisons une bibliothèque Python (geopy) pour effectuer des requêtes à une API de géocodage "Nominatim" d'OpenStreetMap. 
<br>Le script parcourt la liste des villes fournie dans le sujet pour en extraire la latitude et la longitude.

Explications techniques :
<br>Pandas : On l'utilise pour stocker les résultats dans un DataFrame, ce qui facilitera l'export final en fichier ".csv".
<br>Geopy / Nominatim : C'est l'outil qui transforme un nom de ville en coordonnées numériques.
<br>time.sleep(1) : On ajoute une pause d'une seconde entre chaque ville pour respecter les serveurs et éviter d'être banni.
<br>Gestion d'erreurs (try/except) : Si une ville n'est pas trouvée ou si le serveur ne répond pas, le script affiche un message d'erreur et passe à la ville suivante au lieu de s'arrêter complètement.

````python
# =================================================================
# INSTALLATION DES LIBRAIRIES (Environnement)
# =================================================================

# On installe toutes les dépendances nécessaires en une seule fois
%pip install selenium webdriver-manager boto3 sqlalchemy psycopg2-binary plotly geopy -q

#Selenium / Webdriver-manager : Pour mon scraping Booking.
# Boto3 : Pour parler à Amazon S3.
# SQLAlchemy / Psycopg2-binary : Pour envoyer mes données vers ma base de données RDS.
# Plotly : Pour mes futures cartes interactives.
# Geopy : Pour transformer mes noms de villes en coordonnées GPS.

# =================================================================
# IMPORTATIONS DES BIBLIOTHÈQUES
# =================================================================

# --- Standards & Utilitaires ---
import json
import time
import re
from io import StringIO

# --- Manipulation de données ---
import pandas as pd
import numpy as np

# --- Géolocalisation ---
from geopy.geocoders import Nominatim

# --- Scraping Statique & API ---
import requests
from bs4 import BeautifulSoup

# --- Scraping Dynamique (Selenium) ---
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

# --- Cloud & Bases de données (AWS) ---
import os
import boto3
from sqlalchemy import create_engine
from botocore.exceptions import ClientError
from dotenv import load_dotenv

# --- Visualisation Interactive ---
import plotly.express as px
import plotly.graph_objects as go

print("✅ Environnement prêt : Toutes les bibliothèques sont chargées!")
````

````python
# Liste des villes cibles
villes = [
    "Mont Saint Michel", "St Malo", "Bayeux", "Le Havre", "Rouen", "Paris", 
    "Amiens", "Lille", "Strasbourg", "Chateau de Chambord", "Orleans", "Avallon", 
    "Dijon", "Besancon", "Annecy", "Grenoble", "Lyon", "Gorges du Verdon", 
    "Bormes les Mimosas", "Cassis", "Marseille", "Aix en Provence", "Avignon", 
    "Uzès", "Nimes", "Aigues Mortes", "Saintes Maries de la Mer", "Collioure", 
    "Carcassonne", "Ariege", "Toulouse", "Montauban", "Biarritz", "Bayonne", "La Rochelle"
]

# Initialisation du géocodeur
geolocator = Nominatim(user_agent="kayak_project_agent")
villes_data = []

print("Début de la récupération des coordonnées...")

for ville in villes:
    try:
        location = geolocator.geocode(f"{ville}, France")
        if location:
            villes_data.append({
                "city": ville,
                "latitude": location.latitude,
                "longitude": location.longitude
            })
            print(f"✅ {ville} trouvé.")
        else:
            print(f"❌ {ville} non trouvé.")
        time.sleep(1) 
    except Exception as e: 
        print(f"Erreur pour {ville}: {e}")

# Création du Dataframe à partir des données récupérées et ajout d'une colonne d'ID
df_villes = pd.DataFrame(villes_data)

# Insertion d'une colonne 'city_id' en première position, avec des IDs uniques pour chaque ville
df_villes.insert(0, 'city_id', range(1, len(df_villes) + 1))

# Affichage du résultat final
print("\n✅ Résultat final avec IDs :")
print(df_villes.head())

# Sauvegarde pour la suite
df_villes.to_csv("villes_coords.csv", index=False)
````

````python
# Ma clé API
API_KEY = "b73711a24d25a2b2acdd9ac039de9b3b"

# On utilise df_villes (celui avec les IDs) pour faire les appels API et récupérer la météo
weather_results = []
print("Récupération de la météo via API OpenWeather...")

for index, row in df_villes.iterrows():
    # Appel de l'API avec les coordonnées déjà présentes dans notre DataFrame
    url = f"https://api.openweathermap.org/data/2.5/forecast?lat={row['latitude']}&lon={row['longitude']}&appid={API_KEY}&units=metric&lang=fr"
    
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        # On prend le premier créneau de prévision
        first_forecast = data['list'][0]
        
        weather_results.append({
            "city_id": row['city_id'], # On garde l'ID pour la fusion future
            "city": row['city'],
            "temp": first_forecast['main']['temp'],
            "rain_prob": first_forecast.get('pop', 0), # On récupère la probabilité de pluie (0 si absent)
            "weather_description": first_forecast['weather'][0]['description']
        })
        print(f"Météo OK pour {row['city']}")
    else:
        print(f"Erreur API pour {row['city']} (Code {response.status_code})")
    
    # Petite pause pour ne pas saturer l'API (bonne pratique)
    time.sleep(0.2)

# Création du DataFrame Météo
df_meteo_results = pd.DataFrame(weather_results)

# On fusionne sur 'city_id' et 'city' pour garder toutes les infos propres
df_weather_final = pd.merge(
    df_villes, 
    df_meteo_results[['city_id', 'temp', 'rain_prob', 'weather_description']], 
    on="city_id"
)

# Sauvegarde du fichier consolidé
df_weather_final.to_csv("villes_meteo_coords.csv", index=False)

print("\nTERMINÉ ! Le fichier 'villes_meteo_coords.csv' est prêt avec IDs et probabilité de pluie.")
display(df_weather_final.head())
````

### 1.3 Scraping

<br>Pour ce projet, nous allons extraire des données directement depuis le code HTML d'un site web et utiliser la bibliothèque "BeautifulSoup" (pour analyser le code HTML) combinée à "Requests".

Le défi technique :
<br>Booking.com possède des mesures de sécurité contre les robots. Pour réussir l'extraction, nous devons :
<br>Utiliser un User-Agent : C'est une signature envoyée au site pour simuler un navigateur réel (Chrome, Firefox) afin de ne pas être bloqué immédiatement.
<br>Cibler les bonnes balises HTML : Les données sont structurées dans des balises spécifiques (ex: les noms d'hôtels dans des balises "h3" ou des "div" avec l'attribut "data-testid="title"").

Méthodologie :
<br>Le script va parcourir nos 35 villes. Pour chaque ville, il simulera une recherche sur Booking.com et récupérera les noms, les notes et les descriptions des hôtels disponibles.

Explications techniques :
<br>BeautifulSoup(response.text, "html.parser") : Cette ligne transforme le code source de la page (texte brut) en un objet Python navigable comme une arborescence.
<br>data-testid : Les sites modernes utilisent ces attributs stables. C'est plus fiable que les classes CSS qui changent régulièrement.
<br>Boucle sur les villes : La fonction de scraping est intégrée dans une boucle "for" qui traite l'ensemble de notre liste de destinations.

````python
# Liste des 35 villes cibles
cities = [
    "Mont Saint Michel", "St Malo", "Bayeux", "Le Havre", "Rouen", "Paris", "Amiens", 
    "Lille", "Reims", "Nancy", "Strasbourg", "Colmar", "Besançon", "Dijon", "Annecy", 
    "Grenoble", "Lyon", "Gordes", "Avignon", "Arles", "Montpellier", "Sète", 
    "Perpignan", "Carcassonne", "Bayonne", "Biarritz", "Pau", "Albi", "Cahors", 
    "Sarlat la Canéda", "La Roque Gageac", "Rocamadour", "Saintes Maries de la Mer", 
    "Collioure", "Cassis"
]

def scrape_hotels_v1_requests(city_list):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    results = []
    output_file = 'hotels_booking_35_villes.csv'

    print("DÉMARRAGE DU SCRAPING V1 (Méthode Requests)...")
    print("-" * 50)

    for city in city_list:
        url = f"https://www.booking.com/searchresults.fr.html?ss={city}"
        
        try:
            response = requests.get(url, headers=headers, timeout=10)
            
            # Vérification du code HTTP
            if response.status_code == 202:
                print(f"{city} : Code 202 reçu (Requête acceptée mais pas de données).")
                continue
                
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Récupération des cartes d'hôtels
            hotel_cards = soup.find_all('div', {'data-testid': 'property-card'})
            
            if not hotel_cards:
                print(f"❌ {city} : 0 hôtel trouvé (Contenu dynamique non chargé).")
            else:
                print(f"✅ {city} : {len(hotel_cards)} hôtels détectés.")
                
                # Simulation d'extraction pour remplir le CSV (même en V1)
                for card in hotel_cards:
                    results.append({
                        'ville': city,
                        'url_recherche': url,
                        'date_extraction': time.strftime("%Y-%m-%d %H:%M:%S")
                    })

                # --- CHECKPOINT : Sauvegarde après chaque ville ---
                checkpoint_df = pd.DataFrame(results)
                checkpoint_df.to_csv(output_file, index=False)
                print(f"   💾 Sauvegarde intermédiaire effectuée dans {output_file}")

        except Exception as e:
            print(f"{city} : Erreur lors de la requête : {e}")
        
        # Pause pour simuler un comportement humain et éviter le bannissement
        time.sleep(2)

    print("-" * 50)
    print(f"Procédure terminée. Total lignes sauvegardées : {len(results)}")

# Exécution de la fonction
scrape_hotels_v1_requests(cities)
````

Limites de la bibliothèque requests et passage à la bibliothèque Selenium
<br>
Pourquoi ma première tentative a échoué ?
<br> Lors de l'utilisation de la bibliothèque requests, nous avons rencontré deux obstacles majeurs :
<br> Le Code HTTP 202 : Booking.com détecte les requêtes automatisées simples et place la demande dans une "file d'attente" sans renvoyer de données réelles pour protéger son site contre le scraping intensif.
<br> Le Rendu Dynamique (JavaScript) : La structure moderne de Booking repose sur du contenu chargé dynamiquement. 
La librairie "requests" ne récupère que le code HTML source "brut", mais n'exécute pas le JavaScript qui affiche les hôtels et leurs notes.

La solution : L'automatisation par navigateur
<br> Pour contourner ces protections et obtenir les données, nous basculons sur la librairie Selenium.

Contrairement à la bibliothèque précédente, Selenium :
<br> Pilote un véritable navigateur Chrome en mode headless (arrière plan).
<br> Exécute le JavaScript, permettant ainsi l'affichage complet des cartes d'hôtels.
<br> Simule un comportement humain, ce qui permet de passer outre les blocages de type 202.

````python
# Exécution de la nouvelle fonction de scraping avec Selenium

def scrape_hotels_selenium(city): # scrape_hotels_selenium(city)
    chrome_options = Options()
    chrome_options.add_argument("--headless") 
    chrome_options.add_argument("--window-size=1920,1080")
    chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36")

    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)
    
    hotels = []
    city_query = city.replace(" ", "%20")
    url = f"https://www.booking.com/searchresults.fr.html?ss={city_query}"

    try:
        driver.get(url)
        time.sleep(5) 

        # On descend 4 fois de 1000 pixels pour forcer le chargement
        for i in range(4):
            driver.execute_script(f"window.scrollTo(0, {(i+1)*1000});")
            time.sleep(2) # On laisse 2 secondes au site pour charger les nouveaux hôtels

        # On cherche les cartes
        cards = driver.find_elements(By.CSS_SELECTOR, '[data-testid="property-card"]')
        
        # On limite à 20
        for card in cards[:20]:
            try:
                # Nom de l'hôtel
                name = card.find_element(By.CSS_SELECTOR, '[data-testid="title"]').text.strip()

                # URL de l'hôtel
                hotel_url = card.find_element(By.CSS_SELECTOR, '[data-testid="title-link"]').get_attribute('href')

                # Note de l'hôtel (avec sécurité si pas de note)
                try:
                    score_element = card.find_element(By.CSS_SELECTOR, '[data-testid="review-score"]')
                    score_raw = score_element.text
                    match = re.search(r'(\d[,\.]\d)', score_raw)
                    rating = float(match.group(1).replace(',', '.')) if match else 0.0
                except:
                    rating = 0.0

                # Description de l'hôtel
                try:
                    description = card.find_element(By.CSS_SELECTOR, '[data-testid="recommended-units"]').text.strip()
                except:
                    description = "Description non disponible"

                if name:
                    hotels.append({
                        "city": city, 
                        "hotel_name": name, 
                        "url": hotel_url,           
                        "rating": rating, 
                        "description": description  
                    })
            except:
                continue
                
    finally:
        driver.quit()
    
    return hotels
````

````python
# On initialise la liste globale
final_data = []
total_villes = len(df_villes) 
output_file = "hotels_booking_35_villes.csv" # On définit le nom ici

print(f"Lancement du pilotage : 35 villes x 20 hôtels maximum")
print("-" * 50)

# Boucle sur le DataFrame des villes
for index, row in df_villes.iterrows():
    ville = row['city']
    id_ville = row['city_id']
    
    print(f"[{index+1}/{total_villes}] Scraping en cours : {ville}...")
    
    try:
        # On appelle la fonction Selenium
        resultats = scrape_hotels_selenium(ville)
        
        if resultats:
            for hotel in resultats:
                hotel['city_id'] = id_ville 
            
            final_data.extend(resultats)
            print(f"Succès : {len(resultats)} hôtels trouvés.")
            
            # --- LE CHECKPOINT (AJOUTÉ ICI) ---
            # On transforme la liste actuelle en DataFrame et on sauvegarde tout de suite
            checkpoint_df = pd.DataFrame(final_data)
            checkpoint_df.to_csv(output_file, index=False)
            print(f"💾 Checkpoint : {len(final_data)} lignes sauvegardées dans {output_file}")
            
        else:
            print(f"Aucun hôtel trouvé pour {ville}.")
            
    except Exception as e:
        print(f"Erreur sur {ville} : {e}")
    
    # Pause pour laisser respirer le site Booking
    time.sleep(5)

# --- FIN DE LA BOUCLE ---

# On recharge le DataFrame final pour l'affichage
df_hotels = pd.DataFrame(final_data)

print("\n" + "="*30)
print(" SCRAPING TERMINÉ !")
print(f"Total de lignes : {len(df_hotels)}")
print("="*30)

if not df_hotels.empty:
    display(df_hotels.head(20))
````

### 2.1 Nettoyage et agrégation des données (Transformation)

Nettoyage des données (Data Cleaning)
<br> Le scraping de Booking.com peut générer des bruits ou des erreurs. Nous effectuons les actions suivantes :
<br> Filtrage des valeurs manquantes : Suppression des lignes où la note de l'hôtel n'a pas pu être extraite (hôtels sans avis).
<br> Normalisation des formats : Conversion des notes (ex: "8,5") en nombres flottants (Python float) exploitables pour des calculs mathématiques.
<br> Vérification des doublons : On s'assure qu'un même hôtel n'apparaît pas plusieurs fois pour la même ville.

Agrégation par ville (Data Aggregation)
<br> Le dataset issu du scraping contient plusieurs hôtels par destination (68 lignes au total). Pour que ces données soient comparables avec nos prévisions météo (qui n'ont qu'une ligne par ville), nous procédons à une agrégation :
<br> Groupement : Nous utilisons la fonction .groupby('city') sur le nom de la ville.
<br> Calcul de la moyenne : Nous calculons la note moyenne (mean) pour obtenir un "Score Hôtelier" unique par ville.

Pourquoi cette étape est-elle indispensable ?
<br>Qualité des données : Envoyer des données brutes et "sales" sur S3 ou RDS compliquerait les analyses futures.

Réduction de la dimensionnalité : On passe d'une liste d'hôtels éparpillés à un tableau synthétique de 35 lignes (une par ville), prêt à être fusionné avec le dataset météo.

````python
# --- ÉTAPE DE NETTOYAGE ET PRÉPARATION ---

# 1. Chargement des données brutes
# On garde le fichier "hotels_booking_35_villes.csv" tel quel comme archive brute (Raw Data)
df_raw_hotels = pd.read_csv("hotels_booking_35_villes.csv")

# 2. Nettoyage des données
# On crée une copie pour travailler proprement
df_clean_hotels = df_raw_hotels.copy()

# Suppression des lignes sans note (rating)
df_clean_hotels = df_clean_hotels.dropna(subset=['rating'])

# Suppression des doublons (Hôtel identique dans la même ville)
df_clean_hotels = df_clean_hotels.drop_duplicates(subset=['hotel_name', 'city'])

# Gestion des descriptions manquantes (Sécurité pour SQL/RDS)
df_clean_hotels['description'] = df_clean_hotels['description'].fillna("Description non disponible")

# 3. Création du score agrégé par ville
# On calcule la moyenne des notes pour chaque ville en gardant le city_id
df_city_scores = df_clean_hotels.groupby(['city_id', 'city'])['rating'].mean().round(2).reset_index()
df_city_scores.rename(columns={'rating': 'mean_hotel_rating'}, inplace=True)

# 4. Sauvegarde des fichiers finaux
# Fichier 1 : Le dataset "Silver" (nettoyé) - On change le nom pour ne pas écraser le brut
df_clean_hotels.to_csv("hotels_cleaned_full.csv", index=False)

# Fichier 2 : Le dataset agrégé pour la fusion météo
df_city_scores.to_csv("villes_scores_aggregated.csv", index=False)

print(f"✅ Nettoyage terminé avec succès.")
print(f"--------------------------------------------------")
print(f"📊 Statistiques du pipeline :")
print(f"- Données brutes lues    : {len(df_raw_hotels)} lignes")
print(f"- Hôtels uniques sauvés  : {len(df_clean_hotels)} lignes -> 'hotels_cleaned_full.csv'")
print(f"- Villes agrégées        : {len(df_city_scores)} lignes -> 'villes_scores_aggregated.csv'")
print(f"--------------------------------------------------")

display(df_city_scores.head())
````

### 2.2 Fusion avec les données météo

L'importance de la jointure (Merge)
<br>Jusqu'à présent, nos données sont isolées dans plusieurs tableaux (DataFrames). Pour que notre application "Kayak" puisse fonctionner, nous devons lier :
<br>La Géolocalisation (Latitude / Longitude) pour placer les points sur la carte.
<br>La Météo (Température, Humidité, etc.) pour filtrer les meilleures destinations.
<br>La Qualité Hôtelière (Note moyenne) pour affiner le choix final.

Méthodologie : La "Inner Join"
<br>Nous utilisons la fonction pd.merge() de Pandas. La clé de jointure est la colonne "city".
<br>Cohérence : Seules les villes présentes dans les deux datasets (Météo et Hôtels) sont conservées.
<br>Enrichissement : Chaque ligne de notre tableau final contiendra désormais l'intégralité des variables nécessaires à l'analyse.

Résultat final
<br>Le produit de cette fusion est un fichier unique de 35 lignes. Ce fichier est la version "prête à l'emploi" qui sera :
<br>Sauvegardée sur Amazon S3 en tant qu'archive brute.
<br>Injectée dans Amazon RDS pour permettre des requêtes SQL structurées.

````python
# Chargement du fichier météo (créé à l'étape 1.1)
df_weather = pd.read_csv("villes_meteo_coords.csv")

# Fusion avec les scores hôteliers agrégés
# On fait un 'left join' pour garder les 35 villes, même si une n'avait pas d'hôtel
df_final_enrichi = pd.merge(df_weather, df_city_scores, on=['city_id', 'city'], how='left')

# Calcul d'un "Score de Voyage"
# On pondère : 50% météo (température) + 50% qualité hôtelière
# On normalise un peu pour que les chiffres soient cohérents (normalise = on multiplie la note hôtelière par 2 pour la mettre sur une échelle comparable à la température)
df_final_enrichi['travel_score'] = (df_final_enrichi['temp'] * 0.5) + (df_final_enrichi['mean_hotel_rating'] * 2 * 0.5)

# 4. Tri par les meilleures destinations
df_final_enrichi = df_final_enrichi.sort_values(by='travel_score', ascending=False)

# 5. Sauvegarde du fichier final enrichi
df_final_enrichi.to_csv("kayak_final_enriched_data.csv", index=False)

print("Dataset final enrichi créé avec succès !")
print(f"Dimensions : {df_final_enrichi.shape}")
display(df_final_enrichi.head())
````

### 2.3 Stockage sur AWS S3 (Data Lake)
<br> (avec librairie Python boto3)
<br>

### "Note : Ce code est configuré pour AWS, mais l'exécution réelle nécessite un compte actif. Les identifiants ont été retirés pour des raisons de sécurité."

Prérequis : mes identifiants AWS
<br> Avant de lancer le code, nous devons trouver les accès sur notre console AWS :
<br>Dans le service IAM (Identity and Access Management) nous créons un utilisateur (par exemple "Kayak_Project") avec un accès par clé (Programmatic Access).
<br>Nous donnons la permission AmazonS3FullAccess.
<br>Nous récupèrons notre Access Key ID et notre Secret Access Key.

Le code : Envoyer le CSV sur S3
<br>Une fois que nous avons notre DataFrame final (kayak_final_dataset.csv), nous l'envoyons sur S3
<br> (Le format CSV sur S3 est choisi pour sa simplicité, mais entreprise, on utiliserait du Parquet)

Explications :
<br>boto3.Session : C'est comme ouvrir une connexion sécurisée entre notre ordinateur et les serveurs d'Amazon.
<br>StringIO : Au lieu de créer un fichier sur notre ordinateur pour ensuite le télécharger, on transforme le tableau en "flux de texte" directement dans la mémoire vive. C'est plus propre et plus rapide.
<br>Le Bucket : Nous créons le bucket manuellement sur la console AWS avant de lancer le script (on pourrait aussi ajouter une ligne de code pour le créer).

````python
from dotenv import load_dotenv

# --- 1. Chargement et vérification ---
# On utilise '../' pour remonter d'un niveau et sortir du dossier projet
env_path = os.path.join('..', 'config', 'aws.env') 
load_dotenv(dotenv_path=env_path, override=True)

ACCESS_KEY = os.getenv("AWS_ACCESS_KEY_ID")
SECRET_KEY = os.getenv("AWS_SECRET_ACCESS_KEY")
BUCKET_NAME = os.getenv("S3_BUCKET_NAME")
REGION = os.getenv("AWS_REGION", "eu-west-3")

if not ACCESS_KEY or "VOTRE" in ACCESS_KEY: # Petit ajustement de ta condition
    print("ERREUR : Clés absentes ou non configurées dans ../config/aws.env")
else:
    print("✅ Configuration chargée avec succès !")
    # --- 2. Connexion ---
    session = boto3.Session(
        aws_access_key_id=ACCESS_KEY,
        aws_secret_access_key=SECRET_KEY,
        region_name=REGION
    )
    s3 = session.resource('s3')
    s3_client = session.client('s3')

    # --- 3. Vérification/Création du Bucket ---
    try:
        s3_client.head_bucket(Bucket=BUCKET_NAME)
        print(f" Bucket '{BUCKET_NAME}' prêt.")
    except Exception:
        print(f"🔨 Création du bucket '{BUCKET_NAME}'...")
        try:
            s3.create_bucket(
                Bucket=BUCKET_NAME,
                CreateBucketConfiguration={'LocationConstraint': REGION}
            )
        except Exception as e:
            print(f" Impossible de créer le bucket : {e}")

    # --- 4. Fonction d'envoi ---
    def upload_to_s3(file_name, bucket):
        try:
            # On envoie le fichier local avec le même nom sur S3
            s3.Bucket(bucket).upload_file(file_name, file_name)
            print(f" {file_name} envoyé avec succès !")
        except Exception as e:
            print(f" Erreur lors de l'envoi de {file_name} : {e}")

    # --- 5. Liste des fichiers à envoyer ---
    # J'ai mis à jour la liste avec tes nouveaux fichiers nettoyés
    files_to_upload = [
        "villes_meteo_coords.csv",     # Météo
        "hotels_cleaned_full.csv",     # Tes 20 meilleurs hôtels par ville (Nettoyé)
        "villes_scores_aggregated.csv",# La note moyenne par ville (Pour la fusion)
        "kayak_final_enriched_data.csv" # Ton fichier final si déjà créé
    ]

    print(f"\n--- Début de l'upload vers S3 ---")
    for file in files_to_upload:
        if os.path.exists(file):
            upload_to_s3(file, BUCKET_NAME)
        else:
            print(f" Fichier {file} ignoré (non trouvé localement).")
    print(f"--- Fin de la procédure ---\n")
````

Étape suivante l'ETL : extraire les données de S3 pour les injecter dans une base de données RDS (SQL).

Le code : L'ETL vers RDS
<br>Utilisation de SQLAlchemy, bibliothèque Python pour communiquer avec les bases de données.

Explications et Instructions
<br>L'Engine : C'est le moteur qui traduit les commandes Python en langage SQL compréhensible par le serveur Amazon.
<br>to_sql : C'est la méthode Pandas. Elle crée automatiquement la table, définit le type des colonnes (TEXT, FLOAT, INT) et insère toutes les lignes d'un coup.

<br>À une application web (comme Kayak) d'afficher les résultats en temps réel.

Création de deux tables :
<br> Table cities : city_id (Clé primaire), name, latitude, longitude, temp, rain_prob.
<br> Table hotels : hotel_id, city_id (Clé étrangère), name, rating, url.
<br> cette structure permet de faire une Jointure SQL (JOIN) pour mettre à jour la météo tous les jours sans avoir à rescraper les hôtels à chaque fois.

### Le Top 5 des destinations : Carte 1, elle répond à la question "Où aller ?"
<br>

Le concept : Le Top 5 des destinations
<br>Pour définir les "meilleures" destinations, nous devons choisir les villes qui ont la température la plus haute et la plus faible probabilité de pluie.

Le code : Créer la carte interactive

Explications et Instructions
<br>scatter_mapbox : C'est la fonction spécifique de Plotly pour dessiner des points sur une carte géographique.
<br>hover_name et hover_data : C'est ce qui rend la carte interactive. Quand on passe la souris sur un point, une petite étiquette apparaît avec le nom de la ville et les détails météo.
<br>color_continuous_scale : On choisit une échelle de couleurs (du bleu au rouge) pour que l'utilisateur comprenne instantanément où il fait le plus chaud.
<br>mapbox_style="open-street-map" : Cela permet d'avoir un fond de carte gratuit et détaillé sans configuration complexe.

La carte doit répondre immédiatement à la question du client : "Où dois-je partir cette semaine ?".

````python
# On s'assure d'avoir le top 5
top_5_destinations = df_final_enrichi.sort_values(by='travel_score', ascending=False).head(5)

# Création de la carte
fig = px.scatter_mapbox(
    top_5_destinations, 
    lat="latitude",    
    lon="longitude",   
    hover_name="city", 
    hover_data={
        "latitude": False, 
        "longitude": False, 
        "temp": ":.1f", 
        "rain_prob": ":.1f", # Changé en 1f si c'est déjà un chiffre entre 0 et 100
        "mean_hotel_rating": ":.1f"
    },
    color="temp", 
    size="travel_score", 
    color_continuous_scale=px.colors.sequential.YlOrRd, 
    size_max=18,         
    zoom=5,             
    mapbox_style="open-street-map",
    title="Top 5 Destinations : Le meilleur compromis Météo/Hôtels",
    width=800,           
    height=600           
)

# Utilisation de 'autorange' pour que Mapbox cadre parfaitement les 5 points
fig.update_layout(
    margin={"r":10,"t":50,"l":10,"b":10},
    mapbox=dict(
        center=dict(lat=top_5_destinations["latitude"].mean(), 
                    lon=top_5_destinations["longitude"].mean())
    )
)

fig.show()
````

La Formule du Score
<br> Dans le code de fusion que nous avons exécuté, nous avons utilisé cette logique :$$Score = (Météo \times 0.5) + (Qualité\ Hôtelière \times 0.5)$$
<br> Le Facteur Météo ($temp$) : Plus la température est élevée, plus le score monte. Si une ville a 25°C, elle bat une ville qui a 15°C.
<br> Le Facteur Hôtel ($mean\_hotel\_rating$) : On prend la note moyenne des hôtels de la ville. Comme les notes sont sur 10, on les multiplie souvent par un coefficient pour qu'elles aient autant de poids que la température.
<br> L'Équilibre (Pondération) : En mettant 0.5 (50%) sur chaque critère, on s'assure qu'une ville avec une météo de rêve mais des hôtels médiocres ne finisse pas première, et inversement.

Exemple concret :
<br> Ville A (Marseille) : Température 22°C, Note hôtel 8/10.
<br> Ville B (Lille) : Température 12°C, Note hôtel 9.5/10.
<br> Même si les hôtels de Lille sont "mieux notés", le score combiné de Marseille sera plus haut à cause de l'avantage météo.
